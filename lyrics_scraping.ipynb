```python
# Scaping 80s' songs
url_80s = 'https://www.joox.com/th/playlist/vwgMIPd1wV6D3yupd0rUzQ=='

# Send an HTTP GET request to the URL
response_80s = requests.get(url_80s)

# Parse the HTML content of the page
soup_80s = bs(response_80s.text, 'html.parser')

# Find the div element with specified class
div_with_songs_80s = soup_80s.find('div', class_='jsx-2309930405 body')

# Initialize empty lists to store song names and links
songs_80s = []
links_80s = []

if div_with_songs_80s:
    # Find all 'b' elements within the specified 'div' class
    b_elements_80s = div_with_songs_80s.find_all('b')
    
    for b_element_80s in b_elements_80s:
        # Find all 'a' elements within 'b' elements
        a_elements_80s = b_element_80s.find_all('a')  # Find all 'a' elements under 'b' elements
        
        for a_element_80s in a_elements_80s:
            song_name_80s = a_element_80s.text
            song_url_80s = a_element_80s.get('href')
            songs_80s.append(song_name_80s)
            links_80s.append(song_url_80s)

# Create a Pandas DataFrame
song_data_80s = {
    'song_name': songs_80s,
    'song_era': ['80s'] * len(songs_80s),  # Adding '80s' in song_era column
    'song_link': links_80s
}

song_80s = pd.DataFrame(song_data_80s)
song_80s['song_link'] = 'https://www.joox.com' + song_80s['song_link']

print(song_80s)
```

```python
song_80s.to_csv('song_80s.csv', index=False)
```

```python
lyrics_80s = []

# Iterate through each link in the 'song_link' column
for lyrics_80s_link in song_80s['song_link']:
    response_lyrics_80s = requests.get(lyrics_80s_link)
    if response_lyrics_80s.status_code == 200:
        soup_lyrics_80s = bs(response_lyrics_80s.text, 'html.parser')
        div_with_lyrics_80s = soup_lyrics_80s.find('div', {'class': 'jsx-804686892 renderLyrics'})
        if div_with_lyrics_80s:
            p_elements_80s = div_with_lyrics_80s.find_all('p')
            lyric_80s = '\n'.join([p.get_text(strip=True) for p in p_elements_80s])
            lyrics_80s.append(lyric_80s)
        else:
            lyrics_80s.append("Lyrics not found")
    else:
        lyrics_80s.append("Connection error")

# Overwrite the previous 'song_lyrics' column with the updated data
song_80s['song_lyrics'] = lyrics_80s



# Display the updated DataFrame with the lyrics
print(song_80s)
```

```python
song_80s.to_csv('song_lyrics_80s.csv', index=False)
```
